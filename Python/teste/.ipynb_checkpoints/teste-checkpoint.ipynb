```python
import pandas as pd
```

```python
df = pd.read_excel("Organizacao.xlsx", "Resumo", engine="openpyxl")
```

```python
#Retira uma coluna e uma linha desnevcessaria com o drop

dfAtualizado = df.drop([0], axis=0)
dfAtualizado = dfAtualizado.drop(["Unnamed: 0"], axis=1)
```

```python
#Cria duas listas contendo os nomes antigos e novos das colunas, 
# e depois cria um dicionario {ItemAntigo: ItemNovo} com a função ZIP

indicesNovos = []
IndicesAntigos = []
nomesColunas = dfAtualizado.iloc[0]
nomeAtual = dfAtualizado.columns


for x in nomesColunas:
    indicesNovos.append(x)

for item in nomeAtual:
    IndicesAntigos.append(item)


dic_Indice = dict(zip(IndicesAntigos, indicesNovos))
# indicesNovos.clear()
# IndicesAntigos.clear()
```

```python
# Substitui os nomes das colunas pelo dicionario criado, com a função Rename, e deleta a primeira linha, pois estava duplicada
dfAtualizado = dfAtualizado.rename(columns=dic_Indice)
dfAtualizado = dfAtualizado.drop(index=1)
```
