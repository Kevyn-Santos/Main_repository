# Datasets e Series

Uma boa diferenciação a ser feita é: o que são DataSets e o que são Series:

* Datasets são essencialmente tabelas, onde as colunas são chamadas de atributos e as linhas são chamadas de indices
* Series são pedaços da tabela, não é a totalidade ta tabela mas sim uma coluna ou uma linha especifica, ou uma medida feita neste formato

# Seleção de linhas e colunas:

A Seleção de linhas e colunas pode ser feita por meio do índice ou rótulo, e para cada uma possuimos uma fórmula com sintaxe Parecida, Levando uma linha como primeiro argumento e uma coluna como segundo.

## Seleção por índices:
A selção por índices utiliza o método Iloc[], ele espera uma linha como primeiro argumento e uma coluna como segundo argumento, podendo estes serem um intervalo, listas, ou funções. Ele por padrão seleciona a linha primeiro

## Seleção por rótulos
A seleção pro rótulos utiliza o método Loc[], sua sintaxe é igual ao Iloc, porem ele procura pelo nome da linha/Coluna e não pelo seu índice. Ele aceita a busca por listas, tuplas e intervalos

*OBS: Utilizando o método Loc[] o pandas adota os indices(números) das linhas como rótulos, não sendo necessario passa-los entre aspas.*

```python
df.iloc[0:, 0].head() # Localiza todas os registros da linha zero em diante apenas para a coluna zero
```

```python
df.loc[4,'ESTADO'] # Localiza o quarto registro da coluna ESTADO
```

# Informações Gerais do DataFrame:

Algumas informações gerais do DF, como a quantidade de linhas, colunas, e seu tamanho, podem ser adquiridas com o método shape e size.

O primeiro retorna a quantidade de linhas e colunas de um DF como uma tupla, então é possivel desmembrar esta informação em mais de uma variavel.

O segundo retorna o tamanho do DF, que é a quantidade de linhas * a Quantidade de colunas, ou seja, é a quantidade de células do DF. O valor retornado é unico.

Demais informações como a qauntidade de registros e o nome das colunas podem ser adquiridos com os metodos index e columns

```python
tamanho = df.size
print(f'O DataSet possui {tamanho} células')
```

# Atribuição de valores ao dataFrame:

É possivel fazer dois tipos de atribuição de valores ao DataFrame, a atribuição de valores constantes e a atribuição de valores variados.

## Valores constantes:
Caso seja passado um valor unico para a coluna, aquele valor substituirá todos os valores da coluna.

## Valores variados:

Para atribuir diversos valores ao DF é importante que a quantidade de valores seja igual a quantidade de linhas do DF, e é necessario que seja passada uma lista de novos valores. Neste caso podemos usar list comprehension.

```python
dfCopia = df.copy() # df.copy() cria uma cópia do dataframe original, assim evitando que o original seja alterado
dfCopia['ESTADO'] = 0 # Este código atribui o valor '0' pára todos os registros da coluna ESTADO
dfCopia.head()
```

```python
dfCopia2 = df.copy() # É criada a cópia
novosValores = [i for i in range(linhas)] # É usada compreensão de listas para criar uma lista com os novos valores de uma coluna
dfCopia2['ESTADO'] = novosValores # A lista de novos valores é atribuida a coluna

dfCopia2.head()
```

# Criação de colunas no DF:

A criação de colunas em um DF segue uma ideia parecida com a criação de chaves de um dicionario: se não existe, escreva como se fosse editar uma coluna ja existente.

A nova coluna deve conter a mesma quantidade de linhas do DF original.

A criação de novas colunas segue as mesmas lógicas da alteração de colunas ja existintes, ou seja, ela pode conter valores constantes ou variados.

As novas colunas podem ser criadas a partir de colunas ja Existintes no DF

```python
dfCopia3 = df.copy()
listaRegistros = [f"Registro {i}" for i in range(linhas)] # Criação de nova coluna a partir de compreensão de listas
dfCopia3['REGISTROS'] = listaRegistros
dfCopia3['ESTATICO'] ="Teste" # Criação de nova coluna com valor Estatico
dfCopia3['PREÇO MÉDIO REVENDA (Dólar)'] = dfCopia3['PREÇO MÉDIO REVENDA'] * 5.8 # Criação de nova coluna a partir de uma coluna ja existente
dfCopia3.head()
```

# Renomeação de colunas:

A renomeação de colunas é feita com o método Rename, em essencia, é necessario passar um dicionario com o nome antigo como chave e o nome novo como valor, e definindo se o que vai ser modificado é a coluna ou a linha.

É possivel renomear passando valores isolados para cada chave, mapeamento de registros, ou listas

```python
dfrenomeado = df.rename(columns={'Unnamed: 0': 'Índice'})
dfrenomeado.head()
```

#Remoção de colunas

A remoção de colunas pode ser feita com os métodos Del e Drop, mas apenas o Drop será explorado.

O método drop remove uma coluna ou linha baseado em seu índice ou rótulo, é necessario definir se será excluida uma linha ou uma coluna. aceita como parametros valores estaticos, listas ou slices

```python
dfRemovido = dfCopia3.copy()
dfRemovido = dfRemovido.drop(columns='Unnamed: 0') # Retira a coluna com nome 'Unnamed: 0'
dfRemovido.head()
```

```python
dfRemovido = dfRemovido.drop(columns=['REGISTROS', 'ESTATICO']) # Retira as colunas REGISTROS e ESTATICO
dfRemovido.head()
```

```python
dfRemovido = dfRemovido.drop(dfRemovido.iloc[0:0, -1:], axis=1) # Retira a ultima coluna, o -1 indica que é a ultima linha/coluna
dfRemovido.head()
```

# Salvar DF:

Para salvar um Df utilizamos o método to_CSV (ou alguma outra extensão mostrada com o 'to')

Neste caso precisamos passar primeiro o caminho onde o arquivo vai ser salvo, isso é obrigatorio.

Alguns pontos opcionais deste método são:
* podemos utilizar o parametro 'sep' para definir um separador diferente da virgula, que é padrão para arquivos CSV

* Podemos passar o parametro *Index=False* para não criar uma coluna de índice, por padrão esta opção fica ativa, e a coluna gerada não tem nome

No caso de arquivos JSON, passamos o caminho para salvar o arquivo e a orientação(Como ele vai ser salvo), como opcionais temos formatações de numero(ident), data(date_format), e texto(force_Ascii).

o exemplo ficou um pouco mais complicado pois para salvar com a formatação do UTF-8 é necessario abrir o aquivo com a função *with* no modo de escrita, e passando a UTF-8 como tipo de formatação

```python
dfProcessado = dfRemovido.copy()
```

# Filtragem de dados:

A filtragem de dados é feita a partir de comparações diretas de valores. 

## Seleção de uma variavel:
Depois de verificada uma comparação direta, basta armazenar o valor em uma variavel e passa-la no colchetes do DF

```python
dfProcessado['ESTADO'].unique() # o método unique retorna os valores exclusivos de uma coluna
```

```python
CombustivelSãoPaulo = dfProcessado['ESTADO'] == 'SAO PAULO' # retorna uma series dizendo quais valores são verdadeiros e falsos para a condição e os armazena na variavel
```

## Utilizando o método Query:

O método Query mostra do DF com a condicional diretamente, porem é necessario se atentar as comparações de strings. Como o python é case sensitive é necessario utilizar simbolos diferentes na comparação para ele detectar como duas coisas diferentes que estão uma dentro da outra.

O metodo Query não aceita nomes com espaços ou acentos especiais por padrão. O que pode ser feito é mudar o nome das colunas para algo unico ou usar Backsticks (``) para "Burlar" esse problema

```python
dfProcessado.query('ESTADO == "SAO PAULO"').head()
```

```python
dfProcessado.query('`PREÇO MÉDIO REVENDA` >= 1.22').head() # Necessario o uso de backSticks para o query reconhecer a coluna com espaços e acentos
```

## Seleção de mais de uma variavel:

a seleção com mais de uma variavel é feita usando o método query, pois ele comporta operadores condicionais
Mas tambem pode ser feita com duas comparaçoes diferentes, basta colocar cada comparação em um parebteses e "liga-las" com um operador condicional. E então volta para o processo de comparação de uma variavel

& é o operador de AND
| é o operador OR
~ é o operador NOT



```python
dfProcessado.query('`PREÇO MÉDIO REVENDA` >= 1.22 and ESTADO == "SAO PAULO" and PRODUTO == "GNV"').head()
```

```python
precoXestado = (dfProcessado['ESTADO'] == 'SAO PAULO') & (dfProcessado['PREÇO MÉDIO REVENDA'] >= 1.22) & (dfProcessado['PRODUTO'] == 'GNV')
dfProcessado[precoXestado].head()
```
