# Datasets e Series

Uma boa diferenciação a ser feita é: o que são DataSets e o que são Series:

* Datasets são essencialmente tabelas, onde as colunas são chamadas de atributos e as linhas são chamadas de indices
* Series são pedaços da tabela, não é a totalidade da tabela mas sim uma coluna ou uma linha especifica, ou uma medida feita neste formato
* Ou seja, datasets são tabelas enquanto series são colunas ou linhas

# Atribuição de valores ao dataFrame:

É possivel fazer dois tipos de atribuição de valores ao DataFrame, a atribuição de valores constantes e a atribuição de valores variados.

## Valores constantes:
Caso seja passado um valor unico para a coluna, este será o mesmo para todas as linhas da coluna.

## Valores variados:

Para atribuir diversos valores ao DF é importante que a quantidade de valores seja igual a quantidade de linhas do DF, e é necessario que seja passada uma lista de novos valores. Neste caso podemos usar list comprehension.

```python
dfCopia = df.copy() # df.copy() cria uma cópia do dataframe original, assim evitando que o original seja alterado
dfCopia['ESTADO'] = 0 # Este código atribui o valor '0' pára todos os registros da coluna ESTADO
dfCopia.head()
```

```python
dfCopia2 = df.copy()
novosValores = [i for i in range(linhas)] # É usada compreensão de listas para criar uma lista com os novos valores de uma coluna
dfCopia2['ESTADO'] = novosValores # A lista de novos valores é atribuida a coluna

dfCopia2.head()
```

# Criação de colunas no DF:

A criação de colunas em um DF segue uma ideia parecida com a criação de chaves de um dicionario: se não existe, escreva como se fosse editar uma coluna ja existente.

A nova coluna deve conter a mesma quantidade de linhas do DF original.

A criação de novas colunas segue as mesmas lógicas da alteração de colunas ja existintes, ou seja, ela pode conter valores constantes ou variados.

As novas colunas podem ser criadas a partir de colunas ja Existintes no DF

```python
dfCopia3 = df.copy()
listaRegistros = [f"Registro {i}" for i in range(linhas)] # Criação de nova coluna a partir de compreensão de listas
dfCopia3['REGISTROS'] = listaRegistros
dfCopia3['ESTATICO'] ="Teste" # Criação de nova coluna com valor Estatico
dfCopia3['PREÇO MÉDIO REVENDA (Dólar)'] = dfCopia3['PREÇO MÉDIO REVENDA'] * 5.8 # Criação de nova coluna a partir de uma coluna ja existente
dfCopia3.head()
```

## Utilizando o método Query:

O método Query mostra do DF com a condicional diretamente, porem é necessario se atentar as comparações de strings. Como o python é case sensitive é necessario utilizar simbolos diferentes na comparação para ele detectar como duas coisas diferentes que estão uma dentro da outra.

O metodo Query não aceita nomes com espaços ou acentos especiais por padrão. O que pode ser feito é mudar o nome das colunas para algo unico ou usar Backsticks (``) para "Burlar" esse problema

```python
dfProcessado.query('ESTADO == "SAO PAULO"').head()
```

```python
dfProcessado.query('`PREÇO MÉDIO REVENDA` >= 1.22').head() # Necessario o uso de backSticks para o query reconhecer a coluna com espaços e acentos
```

## Duas formas de seleção:

Podemos verificar o DF completo varias vezes, ou a cada verificação criar um DF menor, e fazer as comparações nestes DF menores.

Para o segundo método uma formula seria:

*faz a verificação e guarda em uma variavel -> Passa a variavel no DF -> cria um novo DF com o resultado*


```python
# Verificando o mesmo DF original varias vezes
sel1 = dfProcessado['PRODUTO'] == 'GNV'
sel2 = dfProcessado['ESTADO'] == 'SAO PAULO'
sel3 = dfProcessado['PREÇO MÉDIO REVENDA'] >= 1.22
selFinal = sel1 & sel2 & sel3

dfProcessado[selFinal].head()
```

```python
# A cada verificação, cria um DF menor e verifica sobre o novo DF
# Faz a verificação e guarda em uma variavel -> Passa a variavel no DF -> cria um novo DF com o resultado

CombustivelSãoPaulo = dfProcessado['ESTADO'] == 'SAO PAULO' # Faz a filtragem
CombSP = dfProcessado[CombustivelSãoPaulo] # Cria um novo DF com a filtragem anterior

GNV = CombSP['PRODUTO'] == 'GNV' # Faz a filtragem em cima do novo DF
GNV_sp = CombSP[GNV] # Cria mais um DF menor usando a filtragem anterior

preco_revenda = GNV_sp['PREÇO MÉDIO REVENDA'] >= 1.22 # Faz a verificação em cima do novo DF criado
preco_GNV_SP = GNV_sp[preco_revenda] # Cria um novo DF menor usando a verificação anterior

preco_GNV_SP.head()
```

# Limpeza de dados: 
A limpeza de dados consistem em verificar a integridade e utilidade dos dados, ou seja, trata-los para que não estejam vazios, e para que estejam com os tipos de daods corretos.

Primeiro garantiremos que os dados são do tipo correto. Para verificar os tipos atuais dos dados utilizamos o metodo *.info()*. 

As principais fórmulas para mudar um tipo de dado são:

* to_datetime - converte ua String para data e hora

* to_numeric - converte uma String para inteiro

* astype() - converte um tipo de dado para int, float, bollean, String, ou Category

# Ordenação:

A odenação dos valores é feita especialmente pelo método *sort_values()*, ele pode receber um eixo de atuação como parametro, e pode ser definido se a ordenação será crescente ou decrescente com o parametro *ascending*.

OBS: O método *key* permite passar funções para a ordenação.


```python
notas = pd.DataFrame({
    'nome': ['João', 'Maria', 'José', 'Alice'],
    'idade': [20, 21, 19, 20],
    'nota_final': [5.0, 10.0, 6.0, 10.0]
})
notas
```

```python
notas.sort_values(by='nota_final') # O método recebe um dado pelo qual ele deve ordenar
```

```python
notas.sort_values(by='nota_final', ascending=False) # O parametro ascending=False, retorna os valores em ordem decrescente
```

```python
# Podemos passar listas de ordenação no método, o primeiro elemento da lista será a primeira ordenação, o segundo elemento vai ser uma ordeneção para os valores iguais da primeira ordenaçãoabs
notas.sort_values(by=['nota_final', 'idade'], ascending=[False, True])
# Primeiro ordena de maneira decrescente as notas finais
# para valores iguais, ordena por idade de maneira crescente
```
