```python
import pandas as pd
import numpy as np
import openpyxl
```

```python

Atd1 = pd.read_excel("Atendimentos1112.xlsx", engine="openpyxl")
Atd2 = pd.read_excel("Atendimentos 1712.xlsx", engine="openpyxl")
Atd3 = pd.read_excel("Atendimentos 2312.xlsx", engine="openpyxl", sheet_name="Atendimentos")
```

```python
df = pd.concat([Atd1, Atd2, Atd3], ignore_index=True)
```

```python
dfProcessado = df.drop(df.iloc[0:0,9:], axis=1)
dfProcessado = dfProcessado.drop('Tempo de atendimento', axis=1)
```

```python
dfProcessado = dfProcessado.rename(
    {
    'horario de finalização': 'Horario de Finalização',
    'Data de atendimento': 'Data de Atendimento',
    'Horario de inicio': 'Horario de Inicio'
    }, axis=1
)
```

```python
dfProcessado['Horario de Inicio'] = pd.to_datetime(dfProcessado['Data de Atendimento'].astype(str) + ' ' + dfProcessado['Horario de Inicio'].astype(str))
dfProcessado['Horario de Finalização'] = pd.to_datetime(dfProcessado['Data de Atendimento'].astype(str) + ' ' + dfProcessado['Horario de Finalização'].astype(str))
```

```python
dfProcessado["Duração"] = dfProcessado['Horario de Finalização'] - dfProcessado['Horario de Inicio']
```
