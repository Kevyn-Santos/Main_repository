# Estrutura do Matplotlib:

```python
import matplotlib.pyplot as plt
```

```python
# O matplotlib possui configuarações padrão para varios aspectos dos gráficos, como larguras, posições e cores prédefinidas
# É possivel mudar algumas dessas configurações com o método plt.rcparam['nome do parametro'] = [valor do parametro]. Este método funciona como um dicionario

plt.rcParams['figure.figsize'] = [4,3] # O parametro 'figsize' controla o tamanho do gráfico em polegadas
plt.rcParams["figure.dpi"] = 150 # O parametro 'dpi' muda a resolução do gráfico. O padrão é dpi = 100
plt.rcParams['figure.facecolor'] = 'w' # O parametro 'Facecolor' muda a cor de fundo do gráfico
plt.rcParams['figure.edgecolor'] = 'red' # O parametro 'edgecolor' muda a cor de da borda externa do gráfico

x = np.array([10,100])
y = np.array([50,250])
plt.plot(x,y) # O comando 'plot' cria um gráfico de linha, onde pelo menos x e y precisam ser informados
plt.show() # O comando 'Show' mostra o gráfico ceriado.
```
