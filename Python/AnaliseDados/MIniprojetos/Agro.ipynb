```python
import pandas as pd
import numpy as np

pd.set_option('Display.max_columns', None)
```

# Tratamento dos dados

* Leitura e verificação de dados nulos
* Substituição de valores indevidos por valores nulos
* Remoção de valores nulos
* Conversão dos tipos de dados para float
* Conversão das datas em STR para Datetime

```python
df = pd.read_csv('C:/Users/kevyn/Downloads/Bases de Dados/Kaggle/agricultural_raw_material.csv')
```

```python
# Retorna quantos valores Nulos cada coluna tem
df.isnull().sum()
```

```python
df.head(10)
```

```python
# Remover caracteres indevidos
df = df.replace('%', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('', np.nan)
```

```python
# Remova os valores vazios
df = df.dropna()
```

```python
conversão = list(df.iloc[0:0, 1:])
df[conversão] = df[conversão].astype('float')
df.dtypes
```

```python
df
```

```python
# Convertendo o mês para datetime. Pegue a coluna mês e coloque em maiusculo, depois formate com mês e ano abreviados colocando o ano na frente
df['Month'] = pd.to_datetime(df['Month'].str.upper(), format='%b%y', yearfirst=False)
```

```python
# Set_index define uma coluna como indice no DF, neste caso define a coluna month como indice
df = df.set_index('Month')
df
```
